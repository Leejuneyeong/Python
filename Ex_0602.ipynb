```python
tel={}


while True :
    print("[1]추가 [2]전체조회 [3]삭제 [4]종료")
    menu = int(input(("기능을 선택하세요 >> ")))
    if menu == 1 :
        name = input("이름 : ")
        number = input("연락처 : ")
        tel[name]=number;
        
    elif menu == 2 :
        cnt=1
        for k,v in tel.items() :
            print(f'{cnt}번 {k} {v}')
            cnt+=1
           
    elif menu == 3 :
        cnt=1
        for k,v in tel.items() :
            print(f'{cnt}번 {k} {v}')
            cnt+=1
        name=input('삭제할 이름 >> ')
        
        del tel[name]
        
    elif menu == 4 :
        break
```

```python
def number_minus(num1,num2) :
    result = num1-num2 if num1>num2 else num1-num2
    return result
    #[True] if [Condition] else [False] = 파이썬 삼항연산자
```

```python

num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("첫 번째 정수 입력 >> "))
result = number_minus(num1,num2)
result
```

```python
def cal(num1, num2, op) :
    ''' 두개의 정수로 곱하기나 나누기를 연산하는 함수 '''
    if op == '*':
        result=num1*num2
    elif op == '/':
        result=num1/num2
    return result
```

```python
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))
op = input("연산자 입력(*,/) >> ")
result = cal(num1,num2,op) #shift + tab = 누르면 함수의 정의를 확인할 수 있는 단축키
print("결과 : {}".format(result))
```
