### 3. 연산자 사용하기

```python
# input() -> 결과값이 무조건 문자형태가 된다!

num1 = int(input('첫 번째 정수 입력 : ')) #형변환 해주면 됨 ( 문자 - > 정수)
num2 = int(input('두 번째 정수 입력 : '))

print(num1/num2) # 실수 
print(num1//num2) # 정수만
print(num1%num2) # 나머지
```

```python
#실습
# 1. 두개의 정수 입력 (python 점수, Java 점수)
# 2. 두 과목의 합계 구하기 -> 출력 (합계 : ??)
# 3. 두 과목의 평균 구하기 -> 출력 (평균 : ??)

python = int(input("python 점수 입력 : "))
java = int(input("Java 점수 입력 : "))

sum= python+java
avl= sum/2

print(f'합계 : {sum}')
print(f'평균 : {avl}')

```

```python
time = int(input("시간 입력 >> "))

hour = time // 3600;
minute = time%3600//60
second = time%60

print(f'{hour}시간 {minute}분 {second}초')
```

```python
# 삼항연산자 : (참) a if 조건식 else b (거짓)
# 판단의 기준 -> 점수가 60점 이상 - > 합격 / 아니라면 -> 불합격

score = int(input('점수 입력>>'))

result = "합격" if score >=60 else "불합격"
print(f'{score}점은 {result} 입니다.')
```

```python
num5= int(input('정수 입력 >>'))

result = "짝수" if num5%2==0 else "홀수"

print(f'{num5}은 {result}입니다.')
```
